```python
def grouper(df, idx, col):
    if 80 <= df[col].loc[idx] <= 90:
        return '80-90'
    elif 50 <= df[col].loc[idx] <= 60:
        return '50-60'
    else:
        return 'others'
```

```python
grouper(df, 5, 'temperature')
```

```python
g = df.groupby(lambda x: grouper(df, x, 'temperature'))
g
```

```python
for key, data in g:
    print('key: ',key)
    print(data)
```

```python
# other way

# Define the custom grouping function as a lambda within apply()
df['temp_group'] = df['temperature'].apply(lambda x: '80-90' if 80 <= x <= 90 else '50-60' if 50 <= x <= 60 else 'others')
df
```

```python
# Now, group by the new 'temp_group' column
g = df.groupby('temp_group')
g
```

```python
# Display groups
for key, d in g:
    print(f"Group by Key: {key}\n")
    print(d, "\n")
```
